### 1. Following the function design recipe, define:
  * a function that converts Fahrenheit into Celsius
  * a function that converts kilometers into miles. (Assume there are 1.6 kilometers in a mile.)

```python
# Your code goes here
def f_to_c(f):
    """
    Convert temperature from Fahrenheit to Celsius.

    Args:
    - fahrenheit (float): Temperature in Fahrenheit.

    Returns:
    - float: Temperature converted to Celsius.
    """
    c = (f - 32) * 5 / 9
    return f"{f} fahrenheit is equal to {c:.2f} miles."
print(f_to_c(100))


```

```python
def kilometers_to_miles(kilometers):
    """
    Convert distance from kilometers to miles.

    Args:
    - kilometers (float): Distance in kilometers.

    Returns:
    - float: Distance converted to miles.
    """
    miles = kilometers / 1.6
    return f"{kilometers} kilometers is equal to {miles:.2f} miles."

print(kilometers_to_miles(10))
```

```python
from pickle import TRUE
def is_416_number(number):
    ''' Check whether the number has a 416 area code.
    >>> is_416_number('(416)-555-5555')
    True
    >>> is_416_number('514 416 5555')
    False
    >>> is_416_number('4165554160')
    True
    '''
    # Convert the input into string
    number=str(number)
    # Replace parens w/ nothing
    number = number.replace('(', '')
    number = number.replace(')', '')
    number = number.replace('-','')
    number = number.replace('.','')
    output = number[:3] 
    
    if output== '416':
        return True
    else: 
        return False
print(is_416_number(417))
# False
```

```python
# Second method
import re

def extract_numbers(string):
    """
    Extract all numbers from a string using regular expressions.

    Args:
    - string (str): The input string.

    Returns:
    - list: A list containing all the numbers found in the string.
    """
    # Use regular expression to find all numbers in the string
    string=str(string)
    #numbers = re.findall(r'-?\d+\.?\d*', string)
    # Extract pure integers, ignoring dot '.'
    numbers = re.findall(r'-?\d+', string)
    return ''.join(numbers)
  

from pickle import TRUE
def is_416_number(number):
    ''' Check whether the number has a 416 area code.
    >>> is_416_number('(416)-555-5555')
    True
    >>> is_416_number('514 416 5555')
    False
    >>> is_416_number('4165554160')
    True
    '''
    # extract all digits from a string
    extracted_number_string=extract_numbers(number)
    # this function can extract digits from a string 
    # and decide if it's 10 digit length
    if len(extracted_number_string) != 10:
        return f"Can't extract phone numbers from the string : {number}"
    else: 
        # Check the first 3 numbers in the string
        output = extracted_number_string[:3] 
    
        if output== '416':
            return True
        else: 
            return False
        
print(is_416_number('fefefsdf416fefa'))
# result ==> Can't extract phone numbers from the string : fefefsdf416fefa
```

### 1. What types are involved in the following expressions? What data type will each expression evaluate to?

`8 * 2.5`

```python
8 * 2.5
# 8 ==> Integer; 2.5 ==> Decimal number
# result ==> Float
```

`9 / 2`

```python
9 / 2
# 9 ==> Integer; 2 ==> Integer
# result ==> decimal number
```

`9 // -2`

```python
9 // -2
# 9 ==> Integer; -2 ==> Integer
# result ==> Integer by Floor Division
```

`1.5 * 2 >= 7 - 3`

```python
1.5 * 2 >= 7 - 3
# False == > compare 3 with 4 
```

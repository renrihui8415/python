```python
# Your Part 1 code goes here
# This is a function, which we will learn more about next week. For testing purposes, we will write our code in the function
def anagram_checker(word_a, word_b):
 
  #1 Remove spaces and convert the string to be lowercase
  modified_string_a = word_a.replace(" ","").lower()
  modified_string_b = word_b.replace(" ","").lower()

  #2 Check if the lengths are identical and exclude the empty strings as well
  if len(modified_string_a) != len(modified_string_b) or (len(modified_string_a) ==0 and len(modified_string_b) ==0):
      return False

  #3 Loop through each character in string_a
  for character in modified_string_a:
      if character in modified_string_b:
      # find a match in string_b and remove the match from string_b
        modified_string_b=modified_string_b.replace(character,'',1)
        # replace (,1)means only the first occurrence will be removed
      else:
        # if can't find match in string_b, return false
        return False
  # if loop completes without any issues, return true
  return True

# Write your assertion tests here
def test_anagram_checker():
    # Define test inputs
    word_a = "Slient"
    word_b = "listen"
    
    # Call the function being tested
    result = anagram_checker(word_a, word_b)
    
    # Define the expected result
    expected_result = True
    
    # Assert whether the result matches the expected result
    assert result == expected_result, f"Expected {expected_result}, but got {result}"

# Run the test function
test_anagram_checker()
# No result returned, means expected result returned.
```

```python
# Your Part 3 code goes here
def anagram_checker(words_a, words_b, is_case_sensitive):
    #1 Create a nested function to concatenate the values in the list
    def concatenated_list(list_of_string):
        concatenated_string = ''.join(list_of_string)
        # Remove spaces from the concatenated string
        return concatenated_string.replace(" ","")

    #2 Apply the inner function 
    modified_string_a = concatenated_list(words_a)
    modified_string_b = concatenated_list(words_b)

    #3 decide whether to convert the string to be lowercase
    if is_case_sensitive==False:
        modified_string_a=modified_string_a.lower()
        modified_string_b=modified_string_b.lower()
    
    #4 Check if the lengths are identical and exclude the empty strings as well
    if len(modified_string_a) != len(modified_string_b) or (len(modified_string_a) ==0 and len(modified_string_b) ==0):
        return False

    #5 Loop through each character in string_a
    ## the 'in' operator is case-sensitive, it helps in this situation
    for character in modified_string_a:
        if character in modified_string_b:
        # find a match in string_b and remove the match from string_b
            modified_string_b=modified_string_b.replace(character,'',1)
            # replace (,1)means only the first occurrence will be removed
        else:
            # if can't find match in string_b, return false
            return False
    # if loop completes without any issues, return true
    return True

# Write your assertion tests here
def test_anagram_checker():
    # Define test inputs
    word_a = ["Hello","world"]
    word_b = ["wrooll","held"]
    case_sensitve = True
    
    # Call the function being tested
    result = anagram_checker(word_a, word_b, case_sensitve)
    
    # Define the expected result
    expected_result = False
    
    # Assert whether the result matches the expected result
    assert result == expected_result, f"Expected {expected_result}, but got {result}"

# Run the test function
test_anagram_checker()
# No result returned, means expected result returned.
```

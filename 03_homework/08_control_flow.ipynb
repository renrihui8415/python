### 1. Write a conditional that prints different messages if a bank account balance is below `$`3,000, between `$`3,000 and `$`10,000, or over `$`10,000.

HINT: You can check if a value `x` is between two other values with a condition like `5 <= x <= 10`, but you can also solve this problem by strategically ordering conditions.

````python
# Create a function to convert the input to digit number
def convert_to_digit(input_value):
    if isinstance(input_value, int):
        # If input is already an integer, return it
        return input_value
    elif isinstance(input_value, float):
        # If input is a float, convert it to an integer
        return input_value
    elif isinstance(input_value, str):
        # If input is a string, attempt to convert it to a number
        try:
            # Try converting to integer first
            return int(input_value)
        except ValueError:
            try:
                # If conversion to integer fails, try converting to float
                return float(input_value)
            except ValueError:
                # If conversion to float also fails, return None or handle the error accordingly
                return None
    else:
        # Handle other types of input if necessary
        return None  # or raise an error, depending on your requirements
````

````python
class Balance_Checker:
    # Create a constructor
    def __init__(self, value):
        self.value=value # the value of the balance

    # Create a function to convert the input to digit number
    def convert_to_digit(self):
        if isinstance(self.value, int):
            # If input is already an integer, return it
            return self.value
        elif isinstance(self.value, float):
            # If input is a float, convert it to an integer
            return self.value
        elif isinstance(self.value, str):
            # If input is a string, attempt to convert it to a number
            try:
                # Try converting to integer first
                return int(self.value)
            except ValueError:
                try:
                    # If conversion to integer fails, try converting to float
                    return float(self.value)
                except ValueError:
                    # If conversion to float also fails, return None or handle the error accordingly
                    return None
        else:
            # Handle other types of input if necessary
            return None  # or raise an error, depending on your requirements
    

    # Create 1 function to check the balance
    def check(self):
        converted_balance=self.convert_to_digit()
        if converted_balance is not None:
            if converted_balance<=0:
                print('Your balance is less than 0. Please check.')
            elif converted_balance<3000:
                print('Your balance is under 3000')
            elif converted_balance<=10000:
                print('Your balance is between 3000 and 10000')
            else: 
                print('Your balance is above 10000')
        else: 
            print("Your input is not a valid digit number. Please check." )


````

### 3. Given the list `people`, sort it by people's first name, last name and age. Store the sorted lists as `by_first_name`, `by_last_name`, and `by_age`, respectively.  
`people = [('Mark', 'Harrison', 56), ('Ken', 'Wolseley', 23), ('Emily', 'Robinson', 77)]`

HINT 1: We want our sorting function to return a new list, rather than modifying in place.

HINT 2: `sorted()` accepts a `key` argument. This argument is the name of a function to use when sorting list elements.

````python
# Create a list of 'people'
people = [('Mark', 'Harrison', 56), 
          ('Ken', 'Wolseley', 23), 
          ('Emily', 'Robinson', 77)
          ]
````

````python
# Define a custom sorting function based on the 1st element (First Name) of each tuple
def sort_by_first_name(person):
    return person[0]  # Return the first name of each tuple

# Sort the list of tuples based on the custom sorting function
by_first_name = sorted(people, key=sort_by_first_name)

print(by_first_name)
````

````python
# Define a custom sorting function based on the 2nd element (age) of each tuple
def sort_by_last_name(person):
    return person[1]  # Return the age (2nd element) of each tuple

# Sort the list of tuples based on the custom sorting function
by_last_name = sorted(people, key=sort_by_last_name)

print(by_last_name)
````

````python
# Define a custom sorting function based on the 3rd element (age) of each tuple
def sort_by_age(person):
    return person[2]  # Return the age (3rd element) of each tuple

# Sort the list of tuples based on the custom sorting function
by_age = sorted(people, key=sort_by_age)

print(by_age)
````

### 4. Write a function called `dict_intersect` that takes two dictionaries, `d1` and `d2`, as arguments and returns a set that contains only the keys found in both of the original dictionaries.

HINT: Let's break this down into steps. We need to:
*   Get the keys in `d1`
*   Get the keys in  `d2`
*   Convert them both to sets
*   Find their intersection

HINT: Some useful functions and methods are `keys()`, `set()`, and `intersection()`.

````python
def dic_intersect(d1,d2):
    """
    Return a set containing keys found in both dictionaries.

    Args:
    - d1 (dict): The first dictionary.
    - d2 (dict): The second dictionary.
    """
    d1_keys=set(d1.keys())
    d2_keys=set(d2.keys())
    return d1_keys.intersection(d2_keys)
    # 2nd method:
    # Use set intersection operation to find keys common to both dictionaries
    #return set(d1.keys()) & set(d2.keys())
    

````

````python
d1 = {'a': 1, 'b': 2, 'c': 3}
d2 = {'b': 4, 'c': 5, 'd': 6}
print(dic_intersect(d1,d2))
````

### 5. Write a loop that iterates over the two lists below simultaneously. For each pair of values, print the first number divided by the second. When the program encounters a zero divisor, it should skip the pair without printing anything.  

HINT: We can exit a loop early with `break`.

HINT 2: Remember that we can bundle two lists pairwise with `zip()`.

````python
dividends = [100, 37.5, -12]
divisors = [8, 0, -3]
result_list=[]
for x,y in zip(dividends,divisors):
    if y==0:
        continue
    else:
        result_list.append(x/y)

print(result_list)       
        
````

````python
# 2nd mothod ==> numpy
import numpy as np
dividends = [100, 37.5, -12]
divisors = [8, 0, -3]
# Convert lists to NumPy arrays
x=np.array(dividends)
y=np.array(divisors)

# Suppress the warning for division by zero
np.seterr(divide='ignore', invalid='ignore')
# Perform element-wise division and handle division by zero
result = np.where(y != 0, x / y, np.nan)

print(result)
````
